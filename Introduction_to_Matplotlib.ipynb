```python
#Make plots from a dictonary
butter_price={"yellow":30,
              "peanut":8,
              "almond":12}
fig,ax=plt.subplots()
ax.bar(butter_price.keys(),butter_price.values())
ax.set(title="Robert's Butter Shop",ylabel="Price")
plt.show()
```

```python
#Horizontal Bar Graph 
butter_price={"yellow":30,
              "peanut":8,
              "almond":12}
fig,ax=plt.subplots()
ax.barh(list(butter_price.keys()),list(butter_price.values()))
ax.set(title="Robert's Butter Shop",xlabel="Price",ylabel="Types")
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(10,6))
over_45.plot(kind="scatter",x="age",y="chol",c="target",ax=ax);
ax.set_xlim([45,80])
```

```python
#Making a bit more of a complicated 

fig,ax=plt.subplots(figsize=(10,6))

#plot the Data
scatter=ax.scatter(over_45["age"],over_45["chol"],c=over_45["target"])

#Customize the plot
ax.set(title="Heart Diseas and Cholesterol Levels",xlabel="age",ylabel="Cholesterol");
ax.legend(*scatter.legend_elements(),title="Target");
```

```python
#Adding another plot to existing styled one
fig,(ax0,ax1)=plt.subplots(nrows=2,
                           ncols=1,
                           sharex=True,
                           figsize=(10,8))
#Add data fro Ax0
scatter=ax0.scatter(over_45["age"],
                   over_45["chol"],
                   c=over_45["target"])
#Customize ax0
ax0.set(title="Heart Diseas and Cholesterol Levels",ylabel="Cholesterol");
ax0.legend(*scatter.legend_elements(),title="Target");

#Setup a mean line 

ax0.axhline(y=over_45["chol"].mean(),
            color="b",
            linestyle="--",
            label="Average");
#Add data for ax1
scatter=ax1.scatter(over_45["age"],
                   over_45["thalach"],
                   c=over_45["target"])

#Customize ax1
ax1.set(title="Heart Diseas and Cholesterol Levels",
        xlabel="Age",
        ylabel="Max Heart Rate");
ax1.legend(*scatter.legend_elements(),title="Target");

#Setup a mean line
ax1.axhline(y=over_45["thalach"].mean(),
            color="b",
            linestyle="--",
            label="Average");

#Title
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold");
```

```python
#Changing the xlim and ylim 

#Adding another plot to existing styled one
fig,(ax0,ax1)=plt.subplots(nrows=2,
                           ncols=1,
                           sharex=True,
                           figsize=(10,8))
#Add data fro Ax0
scatter=ax0.scatter(over_45["age"],
                   over_45["chol"],
                   c=over_45["target"])
#Customize ax0
ax0.set(title="Heart Diseas and Cholesterol Levels",ylabel="Cholesterol");
ax0.legend(*scatter.legend_elements(),title="Target");

#Setup a mean line 

ax0.axhline(y=over_45["chol"].mean(),
            color="b",
            linestyle="--",
            label="Average");
#Add data for ax1
scatter=ax1.scatter(over_45["age"],
                   over_45["thalach"],
                   c=over_45["target"])

#Customize ax1
ax1.set(title="Heart Diseas and Cholesterol Levels",
        xlabel="Age",
        ylabel="Max Heart Rate");
ax1.legend(*scatter.legend_elements(),title="Target");

#Setup a mean line
ax1.axhline(y=over_45["thalach"].mean(),
            color="b",
            linestyle="--",
            label="Average");

#Title
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold");
```

```python
#After Changing LIMITS
#Adding another plot to existing styled one
fig,(ax0,ax1)=plt.subplots(nrows=2,
                           ncols=1,
                           sharex=True,
                           figsize=(10,8))
#Add data fro Ax0
scatter=ax0.scatter(over_45["age"],
                   over_45["chol"],
                   c=over_45["target"],
                    cmap="winter")
#Customize ax0
#Set x axis
ax0.set_xlim([50,80])           #XLIM
ax0.set(title="Heart Diseas and Cholesterol Levels",ylabel="Cholesterol");
ax0.legend(*scatter.legend_elements(),title="Target");

#Setup a mean line 

ax0.axhline(y=over_45["chol"].mean(),
            color="b",
            linestyle="--",
            label="Average");
#Add data for ax1
scatter=ax1.scatter(over_45["age"],
                   over_45["thalach"],
                   c=over_45["target"])

#Customize ax1
ax1.set(title="Heart Diseas and Cholesterol Levels",
        xlabel="Age",
        ylabel="Max Heart Rate");
ax1.legend(*scatter.legend_elements(),title="Target");

#Set the y-axis
ax1.set_ylim([60,200])
#Setup a mean line
ax1.axhline(y=over_45["thalach"].mean(),
            color="b",
            linestyle="--",
            label="Average");

#Title
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold");
```
